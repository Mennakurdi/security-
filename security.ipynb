```python
import os
from Crypto.Cipher import DES
from Crypto.Util.Padding import pad, unpad
```

```python

# 1. Setup Environment and Key
key = b'8bytekey'  # DES requires an 8-byte key
iv = os.urandom(8)  # 8-byte initialization vector
```

```python

# 2. Text to encrypt
plaintext = b'Hello DES in CBC Mode'
```

```python

# 3. Pad plaintext to fit 8-byte blocks
padded_text = pad(plaintext, DES.block_size)
```

```python
# 4. Create DES cipher in CBC mode
cipher = DES.new(key, DES.MODE_CBC, iv)
```

```python

# 5. Encrypt the padded plaintext
ciphertext = cipher.encrypt(padded_text)
print(f'Ciphertext: {ciphertext.hex()}')
```

```python
# 6. Decrypting
cipher_dec = DES.new(key, DES.MODE_CBC, iv)
decrypted_text = unpad(cipher_dec.decrypt(ciphertext), DES.block_size)
print(f'Decrypted text: {decrypted_text.decode()}')
```

```python
# 7. Display Intermediate Steps
print(f'IV: {iv.hex()}')
print(f'Padded Text: {padded_text}')
```
